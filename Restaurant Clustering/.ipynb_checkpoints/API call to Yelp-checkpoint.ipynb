```python
key = 'qlan12nw_e_trijd4S2wcHeXLtd9kyO5AX4IDAY5d20V5FB37KHWWD-pMEIf0jhxKF17CRcMW7F4vcxijRkiKhuG9-P0HYqdgX_w6X0RFxCEFAx6efHm8sjjmknKXXYx'
headers = {
        'Authorization': 'Bearer %s' % key,
    }

def get_business_attr(business_id, headers):
    url = 'https://api.yelp.com/v3/businesses/' + business_id
#     print(url)
    response = requests.request('GET', url, headers=headers)
    data = response.json()
    price = data['price']
    review = data['review_count']
    rating = data['rating']
    return [[business_id, price, int(review), rating]]

def main(restaurants, headers):
    df = pd.DataFrame([], columns = ['business_id', 'price', 'review', 'rating'])
    for b_id in tqdm(restaurants['business_id']):
        df_tmp = pd.DataFrame(get_business_attr(b_id, headers), \
                          columns = ['business_id', 'price', 'review', 'rating'])
        df = pd.concat([df, df_tmp])
    return df
```

```python
df = main(restaurants, headers)
df_new = pd.merge(restaurants, df, 
         left_on = 'business_id', right_on = 'business_id', how = 'left')
```

```python
def encode_price(x):
    if x == '$':
        return 1
    elif x == '$$':
        return 2
    elif x == '$$$':
        return 3
    else:
        return 4
    
df_new['expensiveness'] = df_new['price'].apply( lambda x: encode_price(x))
```

```python
df_new.drop(['price','stars', 'review_count'], axis  =1, inplace = True)
```
